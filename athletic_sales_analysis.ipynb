```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
as_2020_df = pd.read_csv("../Challenge 5/Resources/athletic_sales_2020.csv")
as_2021_df = pd.read_csv("../Challenge 5/Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
as_2020_df
```

```python
# Display the 2021 sales DataFrame
as_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
as_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
as_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combine_as_data = pd.concat([as_2020_df, as_2021_df], axis="rows")
combine_as_data1 = combine_as_data.reset_index(drop=True)
combine_as_data1
```

```python
# Check if any values are null.
combine_as_data1.isnull()
```

```python
# Check the data type of each column
combine_as_data1.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
#combine_as_datetime = combine_as_data1.astype({"invoice_date" : datetime64[ns]})
combine_as_data1['invoice_date'] = pd.to_datetime(combine_as_data1['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combine_as_data1.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
total_products_as = combine_as_data1.groupby(["region", "state", "city"])[["units_sold"]].agg('sum')
total_products_as
# Rename the sum to "Total_Products_Sold".
total_products_as_rename = total_products_as.rename(columns={"units_sold": "Total_Products_Sold"})
total_products_as_rename 
# Show the top 5 results.
total_products_as_rename.sort_values(by=["Total_Products_Sold","region", "state", "city"], ascending=False).head(5) 
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_as_pv = pd.pivot_table(combine_as_data1,
                                        index=["region", "state", "city"],
                                        values="units_sold",
                                        aggfunc=sum)
total_products_as_pv
# Rename the "units_sold" column to "Total_Products_Sold"
total_products_as_pv = total_products_as_pv.rename(columns={"units_sold": "Total_Products_Sold"})
total_products_as_pv
# Show the top 5 results.
total_products_as_pv.sort_values(by=["Total_Products_Sold","region", "state", "city"], ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
total_products_as = combine_as_data1.groupby(["region", "state", "city"])[["total_sales"]].agg('sum')
total_products_as
# Optional: Rename the "total_sales" column to "Total Sales"
total_products_as_rename = total_products_as.rename(columns={"total_sales": "Total Sales"})
total_products_as_rename 
# Show the top 5 results.
total_products_as_rename.sort_values(by=["Total Sales","region", "state", "city"], ascending=False).head(5) 
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_products_as_pv = pd.pivot_table(combine_as_data1,
                                        index=["region", "state", "city"],
                                        values="total_sales",
                                        aggfunc=sum)
total_products_as_pv

# Optional: Rename the "total_sales" column to "Total Sales"
total_products_as_pv = total_products_as_pv.rename(columns={"total_sales": "Total Sales"})
total_products_as_pv

# Show the top 5 results.
total_products_as_pv.sort_values(by=["Total Sales","region", "state", "city"], ascending=False).head(5)
```
